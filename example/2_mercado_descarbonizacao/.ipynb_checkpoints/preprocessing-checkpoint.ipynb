```python
import numpy as np
import pandas as pd
```

### Leitura e merge dos dados

```python
CBIO = pd.read_csv('dados/cbio_historico.csv')
CBIO['Data'] = pd.to_datetime(
    CBIO['Data'],
    # dayfirst=True,   # comum em BR (dd/mm/yyyy)
    # errors='coerce',
    format="%Y-%m-%d",
)
cols_float = ['Preço Médio']

CBIO[cols_float] = (
    CBIO[cols_float]
    # .replace({',': '.'}, regex=True)  # caso venha com vírgula decimal
    .astype(float)
)

CBIO_2025 = pd.read_csv('dados/cbio_2025.csv', encoding="latin1", sep=';',index_col=False)
CBIO_2025['Data'] = pd.to_datetime(
    CBIO_2025['Data'],
    # dayfirst=True,   # comum em BR (dd/mm/yyyy)
    # errors='coerce',
    format="%d/%m/%Y",
)
CBIO_2025[cols_float] = (
    CBIO_2025[cols_float]
    .replace({',': '.'}, regex=True)  # caso venha com vírgula decimal
    .astype(float)
)

CBIO_2026 = pd.read_csv('dados/cbio_2026.csv', encoding="latin1", sep=';',index_col=False)
CBIO_2026['Data'] = pd.to_datetime(
    CBIO_2026['Data'],
    # dayfirst=True,   # comum em BR (dd/mm/yyyy)
    # errors='coerce',
    format="%d/%m/%Y",
)
CBIO_2026[cols_float] = (
    CBIO_2026[cols_float]
    .replace({',': '.'}, regex=True)  # caso venha com vírgula decimal
    .astype(float)
)

CBIO = CBIO[['Data',

'Preço Médio']]
CBIO_2026 = CBIO_2026[['Data',

'Preço Médio']]
CBIO_2025 = CBIO_2025[['Data',

'Preço Médio'
]]
def normalize_cols(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
    )
    return df

CBIO = normalize_cols(CBIO)
CBIO_2025 = normalize_cols(CBIO_2025)
CBIO_2026 = normalize_cols(CBIO_2026)
CBIO_full = pd.concat(
    [CBIO, CBIO_2025, CBIO_2026],
    axis=0,
    ignore_index=True,
    sort=False   # não reordena colunas desnecessariamente
)
CBIO_full
CBIO_full.to_csv("CBIO.csv", index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# série estacionária
serie_estavel = np.random.normal(0, 1, 500)

plt.figure(figsize=(12,4))
plt.plot(serie_estavel)
plt.title("Série sem mudança de distribuição")
plt.show()
```

```python
parte1 = np.random.normal(0, 1, 250)
parte2 = np.random.normal(5, 2, 250)

serie_shift = np.concatenate([parte1, parte2])

plt.figure(figsize=(12,4))
plt.plot(serie_shift)
plt.title("Série com Distribution Shift")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))
# plt.hist(parte1, bins=30, alpha=0.6, label='Antes')
# plt.hist(parte2, bins=30, alpha=0.6, label='Depois')
plt.hist(serie_shift, bins=30, alpha=0.6, label='os dois')
plt.legend()
plt.title("Distribuições Antes e Depois")
plt.show()
```

```python
def edf(amostra):
    x = np.sort(amostra)
    y = np.arange(1, len(x)+1) / len(x)
    return x, y

x1, y1 = edf(parte1)
x2, y2 = edf(parte2)
xd, yd = edf(serie_shift)
plt.figure(figsize=(10,5))

plt.step(x1, y1, where='post', label='Antes')
plt.step(x2, y2, where='post', label='Depois')
plt.step(xd, yd, where='post', label='os dois')


plt.title("Funções de Distribuição Empírica")
plt.legend()
plt.show()
```

```python
from scipy.stats import ks_2samp

stat, pvalue = ks_2samp(parte1, parte2)

print("KS Statistic:", stat)
print("p-value:", pvalue)
```

```python
a = np.random.normal(0,1,300)
b = np.random.normal(0,1,300)

stat, pvalue = ks_2samp(a,b)

print(stat)
print(pvalue)
```

```python
t = np.linspace(0,20,500)

s1 = np.sin(t[:250])
s2 = np.sin(3*t[250:])

serie_drift = np.concatenate([s1,s2])

plt.figure(figsize=(12,4))
plt.plot(serie_drift)
plt.axvline(250,color='red',linestyle='--')
plt.title("Concept Drift")
plt.show()
```

```python
t = np.arange(500)

serie_gradual = np.random.normal(
    loc=t*0.01,
    scale=1
)

plt.figure(figsize=(12,4))
plt.plot(serie_gradual)
plt.title("Drift Gradual")
plt.show()
```
