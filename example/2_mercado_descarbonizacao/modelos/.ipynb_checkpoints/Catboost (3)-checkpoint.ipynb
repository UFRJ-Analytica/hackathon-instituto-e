### Carregando os dados

```python
import pickle

# Load it back
with open('data/etanol_series.pkl', 'rb') as f:
    etanol_series = pickle.load(f)

etanol_csv = pd.read_csv("data/etanol_distribuicao.csv")
```

```python
plt.figure(figsize =(15,5))
etanol_series.univariate_component(0).plot(0)
```

```python
'''
etanol_scaler = Scaler()
desvio_scaler = Scaler()

etanol_series_scaled = etanol_scaler.fit_transform(etanol_series.univariate_component(0))
desvio_series_scaled = desvio_scaler.fit_transform(etanol_series.univariate_component(1))
'''
etanol_train , etanol_val = etanol_series_scaled.split_before(0.80)
desvio_train , desvio_val = desvio_series_scaled.split_before(0.85)

train_list_incial = [etanol_train,etanol_val]
val_list = [desvio_train,desvio_val]
```

```python
##Definição de função objetiva
def objective(trial,model_approach):
    global model
    
    params = {
        "lags": trial.suggest_int("lags", 1,12),
        "output_chunk_length": trial.suggest_int("output_chunk_length", 2,100),
        "iterations": 1000,
        "learning_rate": trial.suggest_float("learning_rate", 1e-2, 0.1, log=True),
        "depth": trial.suggest_int("depth", 1, 12),
        "colsample_bylevel": trial.suggest_float("colsample_bylevel", 0.5, 1.0),
        "min_data_in_leaf": trial.suggest_int("min_data_in_leaf", 40, 100),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 6, 10),
        'boosting_type': trial.suggest_categorical('boosting_type', ['Ordered', 'Plain']),
        'eval_metric': "mae",
        'likelihood':'quantile',
        'quantiles':[0.1,0.5,0.9]
    }

    model = XGBModel(**params, multi_models = True)
    
    model.fit(etanol_train)

    # model.save(f'models/Catboost/{model_approach}_trial_{trial.number}.pkl')
 
    predictions = model.predict(n=len(etanol_val)+36)

    score = mape(etanol_val,predictions)
    return score

def callback(study,trial):
    global best_model
    if study.best_trial == trial:
        best_model = model

## Definição do objeto study do optuna
study = optuna.create_study(direction = "minimize")

study.optimize(lambda trial: objective(trial, 'sem_choque'), n_trials=50, callbacks=[callback], show_progress_bar=True)
```

```python
# Fazer previsão estendida com o melhor modelo encontrado
best_preds = best_model.predict(n=len(etanol_val) + 100, num_samples=100)

# Plotar a série original, o período de validação e a previsão
plt.figure(figsize=(15, 5))
etanol_train.plot(label='Treinamento')
etanol_val.plot(label='Validação')
best_preds.plot(label='Previsão')
plt.title('Previsão com XGBModel otimizado via Optuna')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
