```python
import numpy as np
import pandas as pd
from openpyxl import Workbook
from openpyxl.worksheet.worksheet import Worksheet
from openpyxl.drawing.image import Image as OpenPyxlImage
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.utils import get_column_letter
from PIL import Image as PILImage
import io
from typing import Dict, Union, List
from matplotlib.figure import Figure
import timesfm
import matplotlib.pyplot as plt
import seaborn as sns
from darts import TimeSeries
from darts.models import NBEATSModel
from darts.dataprocessing.transformers import Scaler
from darts.metrics import mape
from scipy.signal import stft
```

```python
CBIO = pd.read_csv('data/CBIO.csv')
CBIO
```

```python
# Leitura e pré-processamento SP
df = CBIO[['Data', 'Preço Médio']]
df.columns = ['ds', 'y']
df["unique_id"] = 'BR'
df["ds"] = pd.to_datetime(df["ds"], dayfirst=True, errors="coerce")
df.tail()
```

```python
### pré-process: 
# Criar um novo índice contínuo de dias úteis
full_range = pd.date_range(start=df['ds'].min(),
                           end=df['ds'].max(),
                           freq='B')
uid = df['unique_id'].iloc[0]
df_full = pd.DataFrame({'ds': full_range})
df_full['unique_id'] = uid

# Juntar para ter os valores originais de y com possíveis NaNs
df_full = df_full.merge(df[['ds', 'y']], on='ds', how='left')

# Interpolar os valores de y
df_full['y'] = df_full['y'].interpolate(method='linear')

# Se quiser, preencher no início/fim caso ainda tenha NaN
df_full['y'] = df_full['y'].bfill().ffill()
df_full
```

```python
df_full.to_csv('data/CBIO_data.csv', index=False)
```

## Data Analysis

```python
def plot_df(df, x, y, title="", xlabel='', ylabel='', dpi=200):
    plt.figure(figsize=(18,5), dpi=dpi)
    plt.plot(x, y, color='royalblue')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
    
df_full = df_full[(df_full['ds'] >= '2023-01-01') & (df_full['ds'] <= '2026-12-31')]
    
plot_df(df_full, x=df_full['ds'], y=df_full['y'], title='CBIO price')
```

```python
### growth

x = df_full['ds'].values
y1 = df_full['y'].values

# Plot
fig, ax = plt.subplots(1, 1, figsize=(18,5), dpi= 120)
plt.fill_between(x, y1=y1, y2=-y1, alpha=0.5, linewidth=2, color='seagreen')
plt.ylim(-250, 250)
plt.title('Air Passengers (Two Side View)', fontsize=16)
plt.hlines(y=0, xmin=np.min(df_full['ds']), xmax=np.max(df_full['ds']), linewidth=.5)
plt.show()
```

```python
### Decomposition

from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse

# Multiplicative Decomposition 
multiplicative_decomposition = seasonal_decompose(df_full['y'], model='multiplicative', period=30)

# Additive Decomposition
additive_decomposition = seasonal_decompose(df_full['y'], model='additive', period=30)

# Plot
plt.rcParams.update({'figure.figsize': (11,7.5)})
multiplicative_decomposition.plot().suptitle('Multiplicative Decomposition', fontsize=12)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

additive_decomposition.plot().suptitle('Additive Decomposition', fontsize=12)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```

```python
### Autocorrelation and partial

from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Draw Plot
fig, axes = plt.subplots(1,2,figsize=(16,3), dpi= 100)
plot_acf(df_full['y'].tolist(), lags=700, ax=axes[0])
plot_pacf(df_full['y'].tolist(), lags=10, ax=axes[1])
```

```python
### lags plot

from pandas.plotting import lag_plot
plt.rcParams.update({'ytick.left' : False, 'axes.titlepad':10})

# Plot
fig, axes = plt.subplots(1, 4, figsize=(10,3), sharex=True, sharey=True, dpi=100)
for i, ax in enumerate(axes.flatten()[:4]):
    lag_plot(df_full['y'], lag=((i+2)**2), ax=ax, c='firebrick')
    ax.set_title('Lag ' + str(i+1+i))

fig.suptitle('Lag Plots of Air Passengers', y=1.05)    
plt.show()
```

```python
cbio_series = TimeSeries.from_dataframe(
    CBIO,
    time_col="Data",
    value_cols="Preço Médio",
    fill_missing_dates=True,
    freq="B"   # ajuste: D, W, M, etc.
)
```

## Transforms

### Short Time Fourier Transform

```python
data = cbio_series.values().flatten()
data = pd.Series(data).interpolate(method="linear").to_numpy()
data
```

```python
plt.figure(figsize=(16, 5), dpi=100)
plt.plot(data)
plt.xlabel("Tempo")
plt.ylabel("Preço")
plt.title("Série sintética de preços")
plt.show()
```

```python
f, t, Zxx = stft(data, nperseg=200, noverlap=100)
```

```python
log_magnitude = 20 * np.log10(np.abs(Zxx) + 1e-8)
```

```python
log_magnitude = 10 * np.log10(np.abs(Zxx))

# 4. Plot the Spectrogram
plt.figure(figsize=(10, 5))
plt.pcolormesh(t, f, np.abs(log_magnitude), vmin=0, shading='gouraud', cmap='magma')
plt.title('STFT Spectrogram (Business Day Frequency)')
plt.ylabel('Frequency [Cycles/Day]')
plt.xlabel('Time [Business Days]')
plt.colorbar(label='Magnitude')
plt.tight_layout()
plt.show()
```
